### Conduct a Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(data['Value'])
```

#### The Test Statistic value is greater than the critical value with a confidence level of 5%, which means that we can't reject the null hypothesis. Also the p-value is 1, which means that the test result is not significant. 

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(data)
plt.show()
```

#### All of the lags in the series are out of the confidence level, this means that they are significantly correlated with each other. The data is non-stationary and this supports the Dickey-Fuller test conducted before.

### Differencing the data

```python
data_diff = data - data.shift(1)
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
data_diff.dropna(inplace = True)
```

```python
dickey_fuller(data_diff)
```

### First iteration of ARIMA

```python
# Shape of test
len(test)
```

```python
# Shape of train
len(train)
```

```python
# Import the model
from statsmodels.tsa.api import ARIMA
```

```python
# Model ARIMA (p,d,q) d=stands for how many times the data needed differencing 
model = ARIMA(train, order=(1,1,2))
fitted = model.fit()
print(fitted.summary())

# Forecast
fc =fitted.forecast(155,alpha=0.05) # --> 95% confidence interval(alpha=0.05) / --> 105 is the shape of test

# Make into a pandas series
fc_series = pd.Series(fc, index=test.index) # This creates the forecasted curve
lower_series = pd.Series(fc[0], index=test.index) # Lower bound of confidence interval
upper_series = pd.Series(fc[154], index=test.index) # Upper bound of confidence interval

# Plot
plt.figure(figsize=(12,4),dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=0.05) # This creates the confidence interval
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

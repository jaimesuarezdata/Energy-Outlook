# Unsupervised Machine Learning: Clustering

### Import Libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import plotly.express as px
import plotly.graph_objects as go
from plotly import data
from plotly.subplots import make_subplots
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy import stats
import pylab as pl
```

### Load the files

```python
path = r'C:\Users\Jaime Andres Suarez\Energy Outlook'
df=pd.read_excel(os.path.join(path,'02 Data', 'Transformed Data', 'data_year.xlsx'))
df.head()
```

### Subset the dataframe with only selected columns

```python
df_subset=df[['Emissions','Fossil','GDP','Nuclear','Population','Renewables']]
df_subset.head()
```

### Dataframe basic statistics

```python
df_subset.describe()
```

### Replace empty values with zeros

```python
df_subset=df_subset.fillna(0)
```

### Replace infinite values with zeros

```python
df_subset=df_subset.replace([np.inf,-np.inf],0)
```

### Create a new Scaled DF

```python
df_scaled=pd.DataFrame(df_subset)
```

### Scale the necessary columns 

```python
df_scaled['Population']=stats.zscore(df['Population'])
```

```python
df_scaled['GDP']=stats.zscore(df['GDP'])
```

```python
df_scaled['Emissions']=stats.zscore(df['Emissions'])
```

### Dataframe basic statistics

```python
df_scaled.describe()
```

## Elbow technique

### Define a range of potential clusters

```python
num_cl=range(1,10)
kmeans=[KMeans(n_clusters=i) for i in num_cl]
```

```python
score=[kmeans[i].fit(df_scaled).score(df_scaled) for i in range(len(kmeans))]
score
```

### Plot the elbow curve

```python
pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

#### The optimal count of clusters seems to be 3

### Create the K-means object

```python
kmeans=KMeans(n_clusters=3)
```

### Fit the k-means object to the data

```python
kmeans.fit(df_scaled)
```

### Create a new column with the cluster

```python
df_scaled['cluster']=kmeans.fit_predict(df_scaled)
```

```python
df_scaled.head()
```

### Count of observations in each cluster

```python
df_scaled['cluster'].value_counts()
```

### Plot the clusters for Emissions and Fossil 

```python
plt.figure(figsize=(12,8))
ax=sns.scatterplot(x=df_scaled['Emissions'], y=df_subset['Fossil'], hue=kmeans.labels_,s=100)
ax.grid(False)
plt.xlabel('CO2 Emissions')
plt.ylabel('Fossil energy production')
plt.show()
```

#### The countries that produce the most fossil energy emit the most CO2

### Plot the clusters for GDP and Fossil

```python
plt.figure(figsize=(12,8))
ax=sns.scatterplot(x=df_scaled['GDP'], y=df_subset['Fossil'], hue=kmeans.labels_,s=100)
ax.grid(False)
plt.xlabel('GDP')
plt.ylabel('Fossil energy production')
plt.show()
```

#### The countries with the highest GDP produce the most fossil energy

### Plot the clusters for GDP and Renewable energy production

```python
plt.figure(figsize=(12,8))
ax=sns.scatterplot(x=df_scaled['GDP'], y=df_subset['Renewables'], hue=kmeans.labels_,s=100)
ax.grid(False)
plt.xlabel('GDP')
plt.ylabel('Renewable energy production')
plt.show()
```

#### Most of the countries with the highest GDP produce the most renewable energy; however, there are some countries in the other two GDP groups that produce a significant volume of renewable energy

#### Most of the countries that emit the most CO2 produce the most renewable energy

```python
plt.figure(figsize=(12,8))
ax=sns.scatterplot(x=df_scaled['Fossil'], y=df_subset['Population'], hue=kmeans.labels_,s=100)
ax.grid(False)
plt.xlabel('Fossil energy productions')
plt.ylabel('Population')
plt.show()
```

## Group the countries by cluster

```python
# Rename the clusters
df_scaled.loc[df_scaled['cluster'] == 2, 'cluster'] = 'dark purple'
df_scaled.loc[df_scaled['cluster'] == 1, 'cluster'] = 'purple'
df_scaled.loc[df_scaled['cluster'] == 0, 'cluster'] = 'pink'
```

```python
df_scaled.groupby('cluster').agg({'Emissions':['mean','median'],
                                 'GDP':['mean','median'],
                                 'Fossil':['mean','median'],
                                 'Renewables':['mean','median'],
                                 'Population':['mean','median']})
```

#### Given that the columns had to be scaled down, these statistics do not add much to the final analysis; however, we know that the purple cluster correspond to the countries and years with the highest GDP and consequently the most Fossil energy production, as well as the most CO2 Emissions

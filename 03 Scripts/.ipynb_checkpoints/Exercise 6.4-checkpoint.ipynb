# Exercise 6.4 Supervised Machine Learning: Regression

### import libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import plotly.express as px
import plotly.graph_objects as go
from plotly import data
from plotly.subplots import make_subplots
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

### Load the file

```python
path = r'C:\Users\Jaime Andres Suarez\Energy Outlook'
df=pd.read_excel(os.path.join(path,'02 Data', 'Transformed Data', 'Energy_Data.xlsx'))
df.head()
```

### Melt the DF to transpose year columns into a single column

```python
df_melted=pd.melt(df,id_vars=['Country','Subcategory','Category','Source','OECD','OPEC'],
                  value_vars=[1980,1981, 1982,1983,1984,1985,1986,1987,1988,1989,1990,1991,1992,1993,
                              1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,
                              2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021],var_name='year')
df_melted.head()
```

### Group by Country, Source and year

```python
df_emissions=df_melted.groupby(['Country','Source','year']).agg({'value':'sum'}).reset_index()
df_emissions
```

### Pivot table 

```python
df_pivoted=pd.pivot(df_emissions, values='value',index=['Country','year'],columns="Source").reset_index()
df_pivoted.head()
```

### Derive new column adding Fossil, Nuclear and Renewable energy

```python
df_pivoted['Total Energy production (quad Btu)']=df_pivoted['Fossil']+df_pivoted['Nuclear']+df_pivoted['Renewables']
```

### Derive new column dividing CO2 Emissions by Energy production

```python
df_pivoted['CO2 Emissions (MMtonnes/quad Btu)']=df_pivoted['Emissions']/df_pivoted['Total Energy production (quad Btu)']
df_pivoted.head()
```

```python
df_pivoted.shape
```

### Scatterplot

```python
sns.scatterplot(data=df_pivoted, y='Emissions', x='Total Energy production (quad Btu)')
```

### Correlation coefficient

```python
df_pivoted['Emissions'].corr(df_pivoted['Total Energy production (quad Btu)'])
```

## Hypothesis: is there a relationship between CO2 Emissions and Energy production

### Reshape the variables into Numpy arrays

```python
X=df_pivoted['Total Energy production (quad Btu)'].values.reshape(-1,1)
y=df_pivoted['Emissions'].values.reshape(-1,1)
```

### Split the data into two sets: a training set and a test set

```python
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3, random_state=0)
```

### Create regression object

```python
regression=LinearRegression()
```

### Fit the regression object to the training set

```python
regression.fit(X_train,y_train)
```

### Predict the Emission using the Total Energy Production

```python
y_predicted=regression.predict(X_test)
```

### Create a plot to show the regression line on the test set

```python
plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s=15)
plot_test.plot(X_test, y_predicted,color='red',linewidth=3)
plot_test.title('CO2 Emissions vs Total Energy Production')
plot_test.xlabel('Total Energy Porduction')
plot_test.ylabel('CO2 Emissions')
plot_test.show()
```

#### The line with the predicted values appears to be very approximate to the train values, which shows a strong confidence on the model.

### Check the model performance statistics

#### Create and print objects that contain the model summary statistics

```python
rmse=mean_squared_error(y_test,y_predicted)
r2 = r2_score(y_test, y_predicted)
print('Slope', regression.coef_)
print('Mean squared error: ',rmse)
print('R2 score: ',r2)
```

### Create a dataframe comparing the real and the predicted values of Emissions

```python
data=pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':y_predicted.flatten()})
data.head(30)
```

#### The R2 value shows that there is a strong positive relationship between the two variables. However, the MSE seems highm which makes the model a bit unreliable. Upon comparing the predicted values with the actual values, negative predicted values can be observed, which is impossible, given that there can not be negative CO2 Emissions. The model shows to be inefficient.
